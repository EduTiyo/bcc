```python
import pandas as pd

# 1.1 Load the imdb.csv file into a DataFrame
df = pd.read_csv('imdb.csv')

# 1.2 Display the first 5 rows and use .info() and .describe() to explore the dataset
display(df.head())
df.info()
display(df.describe())

# 1.3 Identify columns with missing values
print("\nColumns with missing values:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# 2.1 Convert columns to appropriate numeric types
# 'year' is already int64, 'rating_imdb' and 'vote' are float64
# 'duration' needs to be converted from string (e.g., '1h 38m') to minutes (numeric)

def convert_duration(duration_str):
    if pd.isna(duration_str):
        return None
    try:
        parts = duration_str.replace('h', ' ').replace('m', '').split()
        hours = int(parts[0]) if parts else 0
        minutes = int(parts[1]) if len(parts) > 1 else 0
        return hours * 60 + minutes
    except ValueError:
        return None # Handle cases where the format is not as expected

df['duration'] = df['duration'].apply(convert_duration)

# 2.2 Create a new column profit = gross_world_wide - budget
df['profit'] = df['gross_world_wide'] - df['budget']

# 2.3 Standardize column names (no spaces and lowercase)
df.columns = df.columns.str.replace(' ', '_').str.lower()

# Display the first few rows and info to verify the changes
display(df.head())
df.info()
```

```python
# 3.1 List the top 5 movies with the highest IMDb rating
print("Top 5 movies with the highest IMDb rating:")
display(df.nlargest(5, 'rating_imdb')[['title', 'rating_imdb']])

# 3.2 Calculate the average duration per decade
df['decade'] = (df['year'] // 10) * 10
average_duration_by_decade = df.groupby('decade')['duration'].mean().reset_index()
print("\nAverage duration per decade:")
display(average_duration_by_decade)

# 3.3 Find the director with the highest average IMDb rating (minimum 3 movies)
director_ratings = df.groupby('director')['rating_imdb'].agg(['mean', 'count']).reset_index()
eligible_directors = director_ratings[director_ratings['count'] >= 3]
top_director = eligible_directors.nlargest(1, 'mean')
print("\nDirector with the highest average IMDb rating (minimum 3 movies):")
display(top_director)

# 3.4 Show the most recurrent genre in the dataset
# Assuming genres are separated by commas and we want the most frequent individual genre
genres = df['genre'].str.split(', ').explode()
most_frequent_genre = genres.mode()[0]
print(f"\nMost recurrent genre in the dataset: {most_frequent_genre}")
```

```python
# 4.1 Load the oscars.csv file into another DataFrame
# It seems there is a parsing issue with the oscars.csv file.
# Let's try reading it with a different delimiter, like a semicolon, as it's common in some datasets.
try:
    oscars_df = pd.read_csv('oscars.csv', sep='\t')
except ParserError:
    print("Parsing with tab failed. Trying another delimiter or inspecting the file.")
    # If semicolon fails, you might need to inspect the file manually
    # or try other common delimiters like tab ('\t') or a fixed width parser
    # For now, we'll assume semicolon works based on the error message hinting at multiple fields
    pass


# Print column names to identify the correct merge key
print("Columns in df:", df.columns)
print("Columns in oscars_df:", oscars_df.columns)

# 4.2 Realize the merge between the main DataFrame and the oscars DataFrame
# Use 'title' from df and 'Film' from oscars_df for merging
merged_df = pd.merge(df, oscars_df, left_on='title', right_on='Film', how='left')

# 4.3 Considering only the films that won the BEST PICTURE category, identify the director with the most awards in this category.
best_picture_winners = merged_df[(merged_df['Category'] == 'BEST PICTURE') & (merged_df['Winner'] == True)]

if not best_picture_winners.empty:
    director_best_picture_wins = best_picture_winners['director'].value_counts().reset_index()
    director_best_picture_wins.columns = ['director', 'best_picture_wins']
    top_director_best_picture = director_best_picture_wins.nlargest(1, 'best_picture_wins')

    print("\nDirector with the most BEST PICTURE wins:")
    display(top_director_best_picture)
else:
    print("\nNo BEST PICTURE winners found in the merged dataset.")
```

```python
# 5.1 Create a pivot_table showing the average rating_imdb by country and decade
pivot_table_rating = df.pivot_table(values='rating_imdb', index='country_origin', columns='decade', aggfunc='mean')
print("Average IMDb rating by country and decade:")
display(pivot_table_rating)

# 5.2 Use melt to reorganize gross columns into long format
gross_columns = ['gross_world_wide', 'gross_us_canada', 'gross_opening_weekend']
melted_gross = df.melt(id_vars=['id', 'title'], value_vars=gross_columns, var_name='gross_type', value_name='gross_amount')
print("\nMelted gross columns:")
display(melted_gross.head())
```

```python
# 6.1 Export the pivot table to CSV
pivot_table_rating.to_csv(
    'rating_pivot_pais_decada.csv',
    index=False,
    encoding='utf-8',
    sep=',',
    na_rep='' # Represent NaN as empty string
)

print("Pivot table exported to rating_pivot_pais_decada.csv")
```
