```python
import pandas as pd

csv = pd.read_csv("imdb.csv")

df = pd.DataFrame(csv)
```

```python
df.head()
df.info()
# df.describe()
# df.isnull()
```

```python
import numpy as np
import pandas as pd

df[['hours', 'minutes']] = df['duration'].str.extract(r'(?:(\d+)\s*h)?\s*(?:(\d+)\s*m)?')
df = df.fillna({'hours': 0, 'minutes': 0})
df[['hours','minutes']] = df[['hours','minutes']].astype(int)
df['duration_min'] = df['hours'] * 60 + df['minutes']

# Fill missing values in 'vote' column with 0 before converting to int
df['vote'] = df['vote'].fillna(0).astype(int)

df = df.rename(
    columns={
        'rating_mpa': 'ratingmpa',
        'rating_imdb': 'ratingimdb',
        'gross_world_wide': 'grossworldwide',
        'gross_us_canada': 'grossuscanada',
        'gross_opening_weekend': 'grossopeningweekend',
        'country_origin': 'countryorigin',
        'filming_location': 'filminglocation',
        'production_company': 'productioncompany'
    }
)

df.head()
```

```python
print('5 maiores imdb:')
print(df['ratingimdb'].nlargest(5), end='\n\n')

print('média de duração por decada:')
i = 1960
while(i < 2015):
  mean_movie = df[(df['year'] >= i) & (df['year'] < i+10)]['duration_min'].mean()
  print(f'{mean_movie:.2f}')
  i+=10

print('diretor com maior média IMDb e com no mínimo 3 filmes')
director_stats = df.groupby('director').agg(
    mean_rating=('ratingimdb', 'mean'),
    movie_count=('ratingimdb', 'count')
)
director_stats = director_stats[director_stats['movie_count'] >= 3]
print(director_stats.nlargest(1, 'mean_rating'))

print('genero mais recorrente')
genre_stats = df.groupby('genre').size().sort_values(ascending=False)
print(genre_stats.nlargest(1))
```

```python
oscars = pd.read_csv("oscars.csv", sep="\t")
oscars_df = pd.DataFrame(oscars)
oscars_df = oscars_df.rename(columns={
    "Film": "title"
})
oscars_df.head()
```

```python
df = pd.merge(df, oscars_df, how="inner", on="title")
```

```python
best_pictures = df[(df['Category'] == 'BEST PICTURE') & (df['Winner'] == True)]

winner_stats = best_pictures.groupby('director').size().reset_index(name="wins")

most_wins_director = winner_stats.nlargest(1, columns='wins')

print(most_wins_director)
```

```python
df['decade'] = (df['year'] // 10) * 10
table = pd.pivot_table(df, values='ratingimdb', index=['decade', 'countryorigin'], aggfunc='mean')
```
